```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./student_performance.csv')
```

```python
df
```

```python
df = df.rename(str.lower, axis=1)
```

```python
df = df.drop(["studentid", "gradeclass"], axis=1)
```

```python
y = df['gpa']
x = df.drop('gpa', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
lr_train_pred = lr.predict(X_train)
lr_test_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

lr_train_mse = mean_squared_error(Y_train, lr_train_pred)
lr_train_mae = mean_absolute_error(Y_train, lr_train_pred)
lr_train_r2 = r2_score(Y_train, lr_train_pred)

lr_test_mse = mean_squared_error(Y_test, lr_test_pred)
lr_test_mae = mean_absolute_error(Y_test, lr_test_pred)
lr_test_r2 = r2_score(Y_test, lr_test_pred)
```

```python
lr_test_mae
```

```python
lr_test_r2
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=1000)
rf.fit(X_train, Y_train)
```

```python
rf_train_pred = rf.predict(X_train)
rf_test_pred = rf.predict(X_test)
```

```python
rf_train_mse = mean_squared_error(Y_train, rf_train_pred)
rf_train_mae = mean_absolute_error(Y_train, rf_train_pred)
rf_train_r2 = r2_score(Y_train, rf_train_pred)

rf_test_mse = mean_squared_error(Y_test, rf_test_pred)
rf_test_mae = mean_absolute_error(Y_test, rf_test_pred)
rf_test_r2 = r2_score(Y_test, rf_test_pred)
```

```python
from sklearn.linear_model import Ridge, RidgeCV
from sklearn.model_selection import RepeatedKFold
```

```python
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
model = RidgeCV(alphas=(0.1, 1.0, 10.0), cv=cv, scoring='neg_mean_absolute_error')
model.fit(X_train, Y_train)
```

```python
ridge_train_pred = model.predict(X_train)
ridge_test_pred = model.predict(X_test)
```

```python
ridge_train_mse = mean_squared_error(Y_train, ridge_train_pred)
ridge_train_mae = mean_absolute_error(Y_train, ridge_train_pred)
ridge_train_r2 = r2_score(Y_train, ridge_train_pred)

ridge_test_mse = mean_squared_error(Y_test, ridge_test_pred)
ridge_test_mae = mean_absolute_error(Y_test, ridge_test_pred)
ridge_test_r2 = r2_score(Y_test, ridge_test_pred)
```
